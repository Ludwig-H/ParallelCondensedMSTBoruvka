# Sanity check: edge accounting on complete graph (N=30)
On doit avoir `sum(len(e) for e in Aretes) == N-1`.

```python
import numpy as np
from parallel_condensed_mst_boruvka import ParallelCondensedMSTBoruvka

rng = np.random.default_rng(0)
N = 30
# complete graph: weights random in [0,1]
U=[]; V=[]; W=[]
for i in range(N):
    for j in range(i+1,N):
        U.append(i); V.append(j); W.append(rng.random())
U=np.array(U, dtype=np.int32); V=np.array(V, dtype=np.int32); W=np.array(W, dtype=np.float64)
Points=[[(i,1.0)] for i in range(N)]
Z, Aretes = ParallelCondensedMSTBoruvka(N,U,V,W,Points,min_cluster_size=2)
total_edges = sum(len(e) for e in Aretes)
print('N=',N,' M=',Z.shape[0],' total_edges=',total_edges)
assert total_edges == N-1, (total_edges, N-1)
print('OK: total_edges == N-1')
```
